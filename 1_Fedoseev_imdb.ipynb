```python
genre_counter = Counter()
for genres in genre_dict.values():
    genre_counter.update(genres)

top_genres = genre_counter.most_common(10)
genres, counts = zip(*top_genres)

colors = plt.cm.Purples(np.linspace(0.9, 0.1, len(counts)))

plt.figure(figsize=(12, 6))
bars = plt.barh(genres, counts, color=colors, edgecolor='black')
plt.ylabel("Genre")
plt.xlabel("Number of films")
plt.title("Top 10 genres by number of films")

for bar, count in zip(bars, counts):
    xval = bar.get_width()
    plt.text(xval + 1, bar.get_y() + bar.get_height() / 2, f'{count}', va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()
```

Самый популярный жанр среди представленных фильмов - драма (всего 58 картин).

```python
genre_gross = Counter()
genre_count = Counter()

for company, genres in genre_dict.items():
    gross = df.loc[df['Series_Title'] == company, 'Gross'].values[0]
    for genre in genres:
        genre_gross[genre] += gross
        genre_count[genre] += 1

average_gross = {genre: genre_gross[genre] / genre_count[genre] for genre in genre_gross}

sorted_avg_gross = sorted(average_gross.items(), key=lambda x: x[1], reverse=True)[:7]
genres, avg_grosses = zip(*sorted_avg_gross)

normalized_colors = np.linspace(0.9, 0.1, len(avg_grosses))
color_map = {gross: plt.cm.Purples(norm_val) for gross, norm_val in zip(avg_grosses, normalized_colors)}
colors = [color_map[gross] for gross in avg_grosses]

plt.figure(figsize=(18, 6))
bars = plt.bar(genres, avg_grosses, color=colors, edgecolor='black')
plt.xlabel("Genre")
plt.ylabel("Average net income")
plt.title("Top 7 genres by average net profit")
plt.xticks(fontsize=10)

for bar, gross in zip(bars, avg_grosses):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{gross:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

Жанр, имеющий наиболшее значение средней чистой прибыли - Family, Sci-Fi.

```python
from wordcloud import WordCloud


x, y = np.ogrid[:800, :800]
mask = (x - 400) ** 2 + (y - 400) ** 2 > 400 ** 2
mask = 255 * mask.astype(int)


countries_text = " ".join(df['Genre'].dropna().tolist())

wordcloud = WordCloud(
    width=800,
    height=800,
    background_color='white',
    max_font_size=100,
    min_font_size=10,
    mask=mask,                 
    contour_color='black',     
    contour_width=1,           
    margin=2,                  
    prefer_horizontal=1,       
    collocations=False         
).generate(countries_text)

def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    darkness = int(255 * (1 - (font_size / 100)))
    return f"rgb({darkness}, {darkness}, {darkness})"  

wordcloud = wordcloud.recolor(color_func=grey_color_func)

plt.figure(figsize=(11, 11))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Genre Word Cloud", pad=20)  

plt.text(320, 850, "Size , Hue: Frequency", fontsize=12, verticalalignment='center')

plt.show()
```

На графике Word Cloud видно, что самый популярный жанр в подборке - драма.

```python
df_scaled = df[['No_of_Votes', 'Gross' ]].apply(lambda x: np.log1p(x))

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_scaled)
plt.title("Comparison of Number of votes and Gross (Log Scaled)")
plt.ylabel("Log scaled values")
plt.xlabel("Columns")
plt.show()
```

На диграммах размаха («ящиках с усами») видно, что у показателя Gross (чистая прибыль) имеется множество выбросов. Т.е. значения доходов с нескольких фильмов далеки от общей выборки. Уровень чистой прибыли различается между собой больше, чем количество голосов (No_of_Votes).

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df[['Released_Year', 'IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of numerical indicators")
plt.show()
```

На Heatmap видно, что наибольшая корреляция между оценками зрителей (IMDB_Rating) и количеством голосов (No_of_Votes) равняется 0,62 (средняя связь). Сильнее всего связано с чистой прибылью (Gross) количество голосов (No_of_Votes). Коэффициент корреляции между этими показателями равен 0,55 (средняя связь). Чистая прибыль (Gross) слабо коррелирует с оценками зрителей (IMDB_Rating), коэффициент всего 0,13. Корреляция между чистой прибылью (Gross) и оценками критиков (Meta_score) равняется -0,015 (слабая отрицательная связь). Между чистой прибылью (Gross) и годом выхода (Released_Year) не наблюдается значимой корреляции, т.к. коэффициент равен 0,23 (слабая связь).

```python
from collections import Counter

# Redefining the code to create the industry dictionary by separating the 'Genre' column values
genre_dict = {}

# Separating industries for each company
for idx, row in df.iterrows():
    film_name = row['Series_Title']
    genres = row['Genre']
    if pd.notna(genres):  
        genres_list = re.split(r"- | & | and | And ", genres)  
        genres_list = [genre.strip() for genre in genres_list]  
    else:
        genres_list = [] 
    genre_dict[film_name] = genres_list

all_genres = []
for genres in genre_dict.values():
    all_genres.extend(genres)

top_10_genres = [genre for genre, count in Counter(all_genres).most_common(10)]


expanded_df = df[['Series_Title', 'Gross']].copy()
expanded_df['Genre'] = expanded_df['Series_Title'].map(genre_dict)
expanded_df = expanded_df.explode('Genre')  

expanded_df_top_10 = expanded_df[expanded_df['Genre'].isin(top_10_genres)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=expanded_df_top_10, y='Genre', x='Gross', orient='h')
plt.xscale('log')  
plt.yticks(rotation=0)

plt.title("Distribution of net profit by 10 leading genres (Log Scale, Horizontal)", pad=20)

plt.xlabel("Gross (Log Scale)")
plt.ylabel("Genre")

plt.tight_layout()
plt.show()
```

Исходя из диаграмм размаха по самым прибыльным жанрам, можно заключить, что значения чистой прибыли в рамках каждого жанра сильно различаются между собой. Об этом говорит длина самих «ящиков» и длина их «усов». Почти у всех жанров имеются выбросы, кроме жанра Animation, Adventure, Comedy.

```python
def plot_top_genre_market_share():
    plt.figure(figsize=(10, 10))
    
    genre_market = df.groupby('Genre')['Gross'].sum().nlargest(10)
    total_market = genre_market.sum()
    market_share = (genre_market / total_market * 100)
    
    colors = sns.color_palette('rocket', n_colors=len(market_share))
    
    textprops = {'fontsize': 12}
    wedgeprops = {'linewidth': 1, 'edgecolor': 'black'}
    
    wedges, texts, autotexts = plt.pie(market_share,
                                       labels=market_share.index,
                                       colors=colors,
                                       autopct='%1.1f%%',
                                       startangle=90,
                                       explode=[0.1 if i == market_share.argmax() else 0 for i in range(len(market_share))],
                                       textprops=textprops,
                                       wedgeprops=wedgeprops)
    
    for i, autotext in enumerate(autotexts):
        if i < 4: 
            autotext.set_color('yellowgreen')
            autotext.set_fontsize(16)  
        else:
            autotext.set_fontsize(12)  

    plt.title('Market share of 10 profitable genres', pad=20)
    plt.axis('equal')
    plt.show()

plot_top_genre_market_share()
```

Самым прибыльным является жанр Action, Adventure, Sci-Fi.
